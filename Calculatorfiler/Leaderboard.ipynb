```python
import json
import matplotlib.pyplot as plt
import numpy as np

# File path to store player rankings
player_rankings_file = 'player_rankings.json'

# Function to load player rankings from file
def load_player_rankings():
    try:
        with open(player_rankings_file, 'r') as file:
            player_rankings = json.load(file)
    except FileNotFoundError:
        # If the file doesn't exist, initialize player_rankings as an empty dictionary
        player_rankings = {}
    return player_rankings

# Function to display the leaderboard
def display_leaderboard():
    player_rankings = load_player_rankings()

    # Sort players based on rankings in descending order
    sorted_players = sorted(player_rankings.items(), key=lambda x: x[1], reverse=True)

    print("\nLeaderboard:")
    rank = 1

    for i, (player, ranking) in enumerate(sorted_players):
        if i == 0 or ranking != sorted_players[i - 1][1]:
            rank = i + 1
        print(f"{rank}. {player}: {ranking}")

# Function to display the leaderboard as a bar chart
def display_leaderboard_bar():
    player_rankings = load_player_rankings()

    # Sort players based on rankings in descending order
    sorted_players = sorted(player_rankings.items(), key=lambda x: x[1], reverse=True)

    names, rankings = zip(*sorted_players)

    plt.figure(figsize=(10, 6))
    plt.bar(range(1, len(names) + 1), rankings, tick_label=names, color='skyblue')
    plt.xlabel('Player')
    plt.ylabel('Ranking')
    plt.title('Leaderboard')
    plt.show()

# Function to display the leaderboard as a scatter plot
def display_leaderboard_scat():
    player_rankings = load_player_rankings()

    # Sort players based on rankings in descending order
    sorted_players = sorted(player_rankings.items(), key=lambda x: x[1], reverse=True)

    names, rankings = zip(*sorted_players)

    plt.figure(figsize=(7, 10))

    # Assign a unique color to each player
    colors = plt.cm.plasma(np.linspace(0, 1, len(names)))

    for i, (name, rank) in enumerate(zip(names, rankings)):
        plt.scatter(2, rank, color=colors[i], marker='x', label=name)

    plt.xlabel('Player')
    plt.ylabel('Ranking')
    plt.title('Leaderboard')
    plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Place the legend outside the plot
    plt.tight_layout()

    # Show the scatter plot
    plt.show()
```

```python
# Display the leaderboard
display_leaderboard()
```

```python
# Display the leaderboard as a bar chart
display_leaderboard_bar()
```

```python
# Display the leaderboard as a scatter plot
display_leaderboard_scat()
```
